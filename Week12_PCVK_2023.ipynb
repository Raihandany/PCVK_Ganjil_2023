```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

def display (img, ):
  fig = plt.figure(figsize=(12,10))
  ax = fig.add_subplot(111)
  ax.imshow(img)
```

```python
cascade_wajah = cv.CascadeClassifier('/content/drive/MyDrive/PCVK_MM/prac_week12/haarcascades/haarcascade_frontalface_alt.xml')

jokowi_img = cv.imread('/content/drive/MyDrive/PCVK_MM/prac_week12/facedet/jokowi.jpg')
kartini_img = cv.imread('/content/drive/MyDrive/PCVK_MM/prac_week12/facedet/kartini.jpg')
kucing_img = cv.imread('/content/drive/MyDrive/PCVK_MM/prac_week12/facedet/kucing.jpg')
mask_img  = cv.imread('/content/drive/MyDrive/PCVK_MM/prac_week12/facedet/mask.png')
jordan_img  = cv.imread('/content/drive/MyDrive/PCVK_MM/prac_week12/facedet/mjordan.jpg')
people_img  = cv.imread('/content/drive/MyDrive/PCVK_MM/prac_week12/facedet/people.jpg')

# Create a list of images
images_arr = [jokowi_img, kartini_img, kucing_img, mask_img, jordan_img, people_img]
plt.figure(figsize = (15,5))

# Process each image separately
for i, img in enumerate(images_arr):
    # Convert the image to grayscale
    rgb_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

    # Detect faces
    roi_wajah = cascade_wajah.detectMultiScale(rgb_img)

    # Draw rectangles around the faces
    for (x, y, w, h) in roi_wajah:
        cv.rectangle(img, (x, y), (x+w, y+h), (255, 255, 255), 3)

    # Plot the image
    plt.subplot(1, len(images_arr), i + 1)
    plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))  # Convert BGR to RGB for correct display
    plt.title(f'Image {i + 1}')

plt.show()

```

```python
kucing_img = cv.imread('/content/drive/MyDrive/PCVK_MM/prac_week12/facedet/kucing.jpg')
kucing_detect = np.zeros_like(kucing_img)

cascade_kucing =  cv.CascadeClassifier('/content/drive/MyDrive/PCVK_MM/prac_week12/haarcascades/haarcascade_frontalcatface.xml')

roi_kucing = cascade_kucing.detectMultiScale(kucing_detect)
for(x,y,w,h) in roi_kucing:
 cv.rectangle(kucing_detect,(x,y),(x+w,y+h),(255,255,255),3)

plt.figure(figsize = (10,10))
plt.subplot(1, len(images_arr), i + 1)
plt.imshow(kucing_img, cv.COLOR_BGR2RGB)
plt.imshow(kucing_detect, cv.COLOR_BGR2RGB)
plt.show()
```
